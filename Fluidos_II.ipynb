# Fluído de una estrella autogravitante, escribo su ecuación de euler.

Planteo la ecuación de Euler:
$$ \frac{D \overline{\rm u}}{Dt} = \frac{1}{\rho} \overline{\rm \nabla} P + \overline{\rm f}^{ ext}$$
Lo que se puede reescribir como:
$$ \frac{D \overline{\rm u}}{Dt} = \frac{∂\overline{\rm u}}{\partial t} + (\overline{\rm u} \cdot \overline{\rm \nabla}) \cdot \overline{\rm u}$$


Tomamos $\overline{\rm u}$ = 0 debido a condiciones hidroestáticas. Por lo tanto podemos llegar a la siguiente conclusión:

$$ \frac{1}{\rho} \overline{\rm \nabla} P = - \overline{\rm f}^{ ext} $$

Siendo $\overline{\rm f}^{ ext} = \overline{\rm f}_{gravedad} $. Utilizando lo indicado en la consigna:

$$\overline{\rm f}_{gravedad} = \rho \cdot  \overline{\rm g}\ \ (1)$$
$$ \nabla \overline{\rm g} = -4 π G ρ\ \ (2)$$

Resuelvo la divergencia de la expresión (2) en r vector (utilizando esféricas debido a la simetria del sistema). Por lo cual llego a la expresión:

$$ \overline{\rm g} = -\frac{ G[4 π  ρ r^{2} dr]}{r^{2}} $$

Llegando la expresión de la masa propuesta por el ejercicio, $ M(r) = -4 π  ρ r^{2} dr$. Utilizando la expreción (1), llegamos a la fuerza gravitatoria:

$$ \frac{1}{\rho} \overline{\rm \nabla} P = -\frac{G M(r)}{r^{2}} \hat{r}$$




# Uso gas politrópico

Primero expandimos la divergencia de la presión expresada en el resultado del primer inciso. Sólo nos interesará la parte en $\hat{r}$ ya que la fuerza gravitatoria (y todas las fuerzas del sistema) están en dicho versor. Además de la antes mencionado también deberemos reemplazar la presión por la ecuación de estado politrópica indicada. Tenemos:
$$ \frac{K}{\rho}  \cdot \frac{\partial ρ^{\frac{n+1}{n}}}{\partial r}= -\frac{G M(r)}{r^{2}} $$

Reescribiendo el diferencial $\partial ρ^{\frac{n+1}{n}}$ (utilizando la expresión diferencial de la presión):

$$ \frac{K}{\rho}  \frac{n+1}{n}\cdot \frac{d ρ}{d r}= -\frac{G M(r)}{r^{2}} $$
La cual al derivar en "r" para simplificar el diferencial/ integral de masa del lado derecho y aplicar la regla de la cadena del lado izquierdo, me queda:

$$ \frac{\partial^2 ρ}{\partial r^2} + \frac{1-n}{n \rho} [\frac{\partial ρ}{\partial r}]^2 + \frac{2}{r} \frac{\partial ρ}{\partial r} + \frac{4\pi G n}{K (n+1)} \rho^{\frac{2n-1}{n}} = 0$$

Obtengo una ecuación diferencial para el campo densidad de la estrella.

# Adimensionalizo

Para empezar a adimensionalizar primero aplico la regla de la cadena a las expresiones de derivadas de $ρ$ (utilizando $ξ = \frac{r}{a}$):

$$ \frac{d}{dr} = \frac{1}{a} \frac{d}{d ξ}$$
$$ \frac{d^2}{dr^2} = \frac{1}{a^2} \frac{d^2}{d ξ^2}$$

Reemplazo con la nueva expresión de la densidad, y debido a la dependencia de theta en ξ tengo que hacer regla de la cadena:

$$ \frac{d ρ}{d ξ} = \rho_{c} n θ^{n}\ \frac{d θ}{d ξ}$$
$$ \frac{d^2 ρ}{d ξ^2} = \rho_{c} [ n θ^{n-1}\ \frac{d^2 θ}{d ξ^2} + n (n-1) \theta^{n-2} (\frac{d θ}{d ξ})^2] $$

Ahora tenemos todas las componentes para adimensionalizar la expresión diferencial de anterior inciso. Hacemos algebra y procuramos que el término de orden más alto (dos en este caso) no quede multiplicado por nada diferente a 1 (nos ayudará para el Runge Kutta). Concluímos en:

$$ \frac{d^2 θ}{d ξ^2} + \frac{1}{ξ} \frac{d θ}{d ξ} + θ^n = 0$$


# Identifico sus condiciones iniciales.

Podemos reescribir la primera ecuacion del inciso (ii) como...

$$ \frac{K}{\rho} \frac{\partial \rho^{\frac{1}{n}+1}}{\partial r} = (n+1)k \frac{\partial \rho^{1/n}}{\partial r} = -\frac{G \, M(r)}{r^2} $$

donde la primera igualdad se obtiene jugando con la derivada parcial, ahora reemplazamos $\rho = \rho_o \,\theta^n$

$$ (n+1)k \frac{\partial (\rho_o^{1/n} \theta)}{\partial r} = -\frac{G \, M(r)}{r^2} $$

pasando las constantes hacia el lado derecho

$$ \frac{\partial \theta}{\partial r} = -\frac{G}{(n+1)k \rho_o^{1/n}} \frac{M(r)}{r^2} $$

notamos que $\frac{\partial \theta}{\partial r} < 0 $ por lo tanto $\theta(r)$ es máxima cuando $r=0$ (y por lo tanto $\rho$ también), esto nos permite definir el $\rho_{max}(r=0) = \rho_0$ y por lo tanto $\theta_{max}(r=0)=1$ . Ahora busquemos averiguar el valor de $\frac{\partial \theta}{\partial r}(r=0)$ para eso tomemos límite en la expresion anterior (reemplazo la constante por un $\lambda$).

$$ \frac{\partial \theta}{\partial r}(r=0) =\lim_{r\to 0} \lambda \frac{M(r)}{r^2} = \lim_{r\to 0} \lambda \frac{\int\rho 4\pi r^2 dr}{r^2} = \lim_{r\to 0} \lambda \frac{\rho 4\pi r^2 dr}{2r} = 0 $$

donde en la ultima igualdad usamos L'Hopital y el TFCI. Así es como llegamos a las CIs:

$$ \theta_{max}(r=0)=1 ,\;\;\; \frac{\partial \theta}{\partial r}(r=0) = 0 $$

```python
import numpy as np
import matplotlib.pyplot as plt

def rk4(x_i, frk, t_i, h, a):
    k_1 = frk(t_i, x_i, a)
    k_2 = frk(t_i + h/2, x_i + k_1 * h/2, a)
    k_3 = frk(t_i + h/2, x_i + k_2 * h/2, a)
    k_4 = frk(t_i + h, x_i + k_3 * h, a)
    return x_i + (h/6)*(k_1 + 2*k_2 + 2*k_3 + k_4)

def estrella(e, u, params):
    x, v = u
    n = params
    x_prima = v
    v_prima = -(2/e) * v - x**n
    return np.array([x_prima, v_prima])

def int_estrella(u_0, estrella, t_f, dt, m_params):
    m_X = [[] for _ in range(len(m_params))]
    m_V = [[] for _ in range(len(m_params))]

    for p in range(len(m_params)):
        n = m_params[p].astype(float)
        u = u_0
        for i in t_f:
            m_X[p].append(u[0])
            m_V[p].append(u[1])
            u = rk4(u, estrella, i, dt, (n))

    m_X = np.array(m_X)
    m_V = np.array(m_V)

    variables_x = {}
    variables_v = {}

    for i in range(len(m_params)):
        pos = "x" + str(i)
        vel = "v" + str(i)
        variables_x[pos] = m_X[i]
        variables_v[vel] = m_V[i]

    return variables_x, variables_v

# CI y Parámetros
u_0 = np.array([1,0])
m_params = np.arange(1, 11, dtype=float)

de = 0.1
e = [1e-16, 10000]
e_f = np.arange(e[0], e[1], de)



variables_x, variables_v = int_estrella(u_0, estrella, e_f, de, m_params)

colores = ['#FF5733', '#3498DB', '#1ABC9C', '#F1C40F', '#E74C3C', '#9B59B6', '#FD9A02', '#964B00', '#00FFFF', '#808080']

plt.figure(figsize=(25, 8))
plt.suptitle(r'Análisis estrella autogravitante, densidad vs $\xi$: n:[1,10]', fontsize=16)

for i in range(len(m_params)):
    n = i + 1
    plt.subplot(2, 5, n)
    plt.plot(e_f, variables_x[f'x{i}'], label=f'$n = {n}$', color=colores[i])
    plt.xlabel(r'Radio [$\xi$]', fontsize=15)
    plt.ylabel(r'Densidad [$\theta$]', fontsize=15)
    plt.legend(fontsize=7)
    plt.grid(linestyle='--')
    plt.grid(which='minor', linestyle=':', linewidth='0.1', color='black')

plt.tight_layout()
plt.show()
```
